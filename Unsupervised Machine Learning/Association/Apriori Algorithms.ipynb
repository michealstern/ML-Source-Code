```python
# Collect data without NaN value:

market = []
for i in range(0,data.shape[0]):  # rows:
    customer = []
    for j in data.columns:        # columns:
       # print(data[j][i])
       if  type(data[j][i]) == str:
           customer.append(data[j][i])
    market.append(customer)
```

```python
# To convert Single list:
import numpy as np
l = []
for i in market:
    for j in i:
       # print(j)
        l.append(j)
```

```python
# l
```

```python
# To find Most frequently selling price:

import collections

collections.Counter(l)
```

```python
p = collections.Counter(l)
```

```python
p.keys()
```

```python
p.values()
```

```python
d = {"Item Name":p.keys(),"values":p.values()}
```

```python
pd.set_option("display.max_rows",500)
pd.DataFrame(d).sort_values(by=["values"],ascending=False)
```

# Convert data into Metrix

```python
# !pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
tr.transform(market)
```

```python
df = pd.DataFrame(tr.transform(market),columns=tr.columns_)
```

# Apriori Algorithms

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df,min_support=0.05,use_colnames=True).sort_values(by=["support"])
```
