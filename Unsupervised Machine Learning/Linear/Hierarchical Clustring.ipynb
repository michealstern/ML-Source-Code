# 1. Agglomerative Hierarchical

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("iris")
data.drop(columns=["species"], inplace=True)
data.head(3)
```

```python
sns.pairplot(data=data)
plt.show()
```

```python
# pip install SciPy
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
sc.dendrogram(sc.linkage(data,method="single",metric="euclidean"))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac =  AgglomerativeClustering(n_clusters=2,linkage='single')
data["Predict"] = ac.fit_predict(data)
```

```python
data.head(3)
```

```python
sns.pairplot(data=data, hue="Predict")
plt.show()
```
