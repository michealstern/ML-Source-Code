```python
globalWarming_df = pd.read_csv("C:\\Users\\arman\\Downloads\\Who_is_responsible_for_global_warming.csv")
globalWarming_df
```

```python
globalWarming_df = globalWarming_df.drop(columns=["Country Code", "Indicator Name", "Indicator Code"], axis = 1).set_index("Country Name")
globalWarming_df.head(6)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, cmap = "coolwarm", annot = True)
```

```python
plt.figure(figsize=(16,9))
annot_kws = {"fontsize":10,
             "fontstyle":"italic",
             'color':'k',
             'alpha':0.9,
             'rotation':'vertical',
             'verticalalignment':'center',
             'backgroundcolor':'w'}
sns.heatmap(globalWarming_df, cmap = "coolwarm", annot = True, annot_kws=annot_kws)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, cmap = "coolwarm", annot = True, linewidths=4)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, cmap = "coolwarm", annot = True, linewidths=4, linecolor='k')
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, cbar=False)
```

```python
plt.figure(figsize=(16,9))

sns.heatmap(globalWarming_df, cbar=False, xticklabels=False)
```

```python
plt.figure(figsize=(16,9))
cbar_kws = {"orientation":"horizontal",
            "shrink":1,
            "extend":'min',
            'extendfrac':0.1,
            'ticks':np.arange(0,22),
            "drawedges":True}
sns.heatmap(globalWarming_df, cbar_kws = cbar_kws)
```

```python
globalWarming_df.corr()
```

```python
plt.figure(figsize=(16,9))
ax = sns.heatmap(globalWarming_df.corr(), annot=True, linewidths=3)
ax.tick_params(size=10, color='w', labelsize=10, labelcolor='w')
plt.title("heatmap of 'who is responsible for global warming' ", fontsize=25)
plt.show()
```
