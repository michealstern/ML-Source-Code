```python
import seaborn as sns
import numpy as np
```

```python
# Load dataset from GitHub seaborn repository

tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.barplot(x = tips_df.total_bill)   #In dataframe accessing the value of total_bill.
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill)
```

```python
sns.barplot(x = tips_df.day, y = tips_df.total_bill, hue = tips_df.sex)
```

```python
# Data =  , use for simplication method

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df)
```

```python
order = ['Sun', 'Thur', 'Fri', 'Sat']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, order = order)
```

```python
hue_order = ['Female', 'Male']

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, hue_order = hue_order)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, estimator = np.max)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, errorbar=('ci', 20))
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, estimator = np.max,
            n_boot = 2)
```

```python
# Error 

sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, orient = 'h')
```

```python
sns.barplot(y = 'day', x = 'total_bill', hue = 'sex', data = tips_df,)
```

```python
sns.barplot(x = 'total_bill', y = 'day', hue = 'sex', data = tips_df,
            orient = 'h')
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, palette = 'magma')
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, saturation = 0.3)
```

```python
sns.barplot(x = 'day', y = 'total_bill', hue = 'sex', data = tips_df, err_kws={'color': '0.5'})
```

```python
sns.set()
kwargs = {'alpha':0.9, 'linestyle':':', 'linewidth':5, 'edgecolor':'k'}

sns.barplot(x = 'day', y = 'total_bill', data = tips_df, **kwargs)
```
