```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
sns.histplot(tips_df["size"])
plt.show()
```

```python
sns.histplot(tips_df["tip"])
plt.show()
```

```python
sns.histplot(tips_df["total_bill"])
plt.show()
```

```python
sns.histplot(tips_df["total_bill"], bins=100)
plt.show()
```

```python
sns.distplot(tips_df["total_bill"], hist = True)
plt.show()
```

```python
sns.distplot(tips_df["total_bill"], kde=False)
plt.show()
```

```python
sns.distplot(tips_df["total_bill"],rug=True)
plt.show()
```

```python
sns.distplot(tips_df["total_bill"], fit=norm)
plt.show()
```

```python
sns.distplot(tips_df["total_bill"], color="r")
plt.show()
```

```python
sns.distplot(tips_df["total_bill"], color="r", label="Total Bill")

plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
sns.set()
sns.distplot(tips_df["total_bill"], color="r", label="Total Bill")

plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
bins = [1,5,10,15,20,25,30,35,40,45,50,55]

sns.set()
sns.distplot(tips_df["total_bill"], color="r", bins = bins)

plt.xticks(bins)

plt.title("Histogram of Total Bill")

plt.show()
```
