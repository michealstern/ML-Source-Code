```python
#Shows Total missing values from columns.
df.isnull().sum()
```

```python
#Shows Total missing values from columns in graphics.
plt.figure(figsize=(25,25))
sns.heatmap(df.isnull())
```

```python
# Missing values shows in precentage(%).
null_var = df.isnull().sum()/df.shape[0] *100
null_var
```

```python
#Remove data from dataset which is missing 
drop_columns  = null_var[null_var >20].keys()
drop_columns
```

```python
# Remove columns.
df2_drop_clm=df.drop(columns=drop_columns)
```

```python
df2_drop_clm.shape
```

```python
#Check dataset
sns.heatmap(df2_drop_clm.isnull())
```

```python
#By default its delay NaN value and creat new dataframe
df3 = df2_drop_clm.dropna()
```

```python
df3.shape
```

```python
#Check dataset
sns.heatmap(df3.isnull())
```

```python
#Check missing value .
df3.isnull().sum().sum()
```

```python
# Combination  of Missing value & Clean dataframe distplot
sns.histplot(df['CRIM'])
sns.histplot(df3['CRIM'])
```

```python
num_var = ['CRIM', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO',
       'LSTAT', 'MEDV']
plt.figure(figsize=(25,25))       
for i,var in enumerate(num_var):
    plt.subplot(4,3,i+1)  # 4 rows, 3 columns.
    sns.histplot(df[var], bins=20)   #Original dataset
    sns.histplot(df3[var], bins=20)  # Clean dataset
```
