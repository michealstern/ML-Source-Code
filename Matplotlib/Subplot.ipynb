```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)

plt.show()
```

```python
plt.pie([1])
plt.pie([1,2])
plt.pie([1,2,3])
plt.pie([1,2,3,4])

plt.show()
```

```python
plt.subplot(2,2,1)
plt.pie([1])

plt.subplot(2,2,2)
plt.pie([1,2])

plt.subplot(2,2,3)
plt.pie([1,2,3])

plt.subplot(2,2,4)
plt.pie([1,2,3,4])

plt.show()
```

```python
plt.figure(figsize=(16,9))

plt.subplot(3,2,1)
plt.subplot(3,2,2)
plt.subplot(3,2,3)
plt.subplot(3,2,4)
plt.subplot(3,2,5)
plt.subplot(3,2,6)

plt.show()
```
