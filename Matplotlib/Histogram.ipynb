```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age = np.random.randint(18,45, (100))
py_students_age = np.random.randint(15,40, (100))

print(ml_students_age)
print(py_students_age)
```

```python
plt.hist(ml_students_age)

plt.title("ML Students age histograms")
plt.xlabel("Students age")
plt.ylabel("No. Students")
plt.show()
```

```python
bins = [15,20,25,30,35,40]
plt.hist(ml_students_age, bins, rwidth=0.8, orientation='vertical', color='red', label='ML Student')

plt.title("ML Students age histograms")
plt.xlabel("Students age")
plt.ylabel("No. Students")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40]

plt.figure(figsize=(16,9))  # Increase graph size
plt.hist(ml_students_age, bins, rwidth=0.8, histtype='bar', 
         orientation='vertical', color='red', label='ML Student')

plt.hist(py_students_age, bins, rwidth=0.8, histtype='bar', 
         orientation='vertical', color='m', label='Python Student')

plt.title("ML Students age histograms")
plt.xlabel("Students age")
plt.ylabel("No. Students")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40]

plt.figure(figsize=(16,9))  # Increase graph size
plt.hist([ml_students_age, py_students_age], bins, rwidth=0.8, histtype='bar', 
         orientation='vertical', color=['red', 'm'], label=['ML Student', 'Python Student'])

#plt.hist(py_students_age, bins, rwidth=0.8, histtype='bar', 
#         orientation='vertical', color='m', label='Python Student')

plt.title("ML Students age histograms")
plt.xlabel("Students age")
plt.ylabel("No. Students")
plt.legend()
plt.show()
```
