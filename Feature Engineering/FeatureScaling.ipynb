# 1. Feature Scaling

## 1.1 Standardization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset("titanic")
df.head(3)
```

```python
df.isnull().sum()
```

```python
df2 = df[['survived',"pclass",'age','parch']]
df2.head()
```

```python
#Data Clean
df3 = df2.fillna(df2.mean())
```

```python
df3.isnull().sum()
```

```python
df3.describe()
```

```python
sns.distplot(df3["age"])
plt.show()
```

```python
# Create a distribution plot with KDE for each column

columns = ['survived',"pclass",'age','parch']

plt.figure(figsize=(10,5))
for column in columns:
    sns.distplot(df3[column], kde=True)  
    plt.title(f'Distribution of {column}')
    plt.show()  # Display each plot one by one
```

```python
sc = StandardScaler()
sc.fit(df3[['age']])
```

```python
sc.transform(df3[['age']])
```

```python
df3['age_sc'] = pd.DataFrame(sc.transform(df3[['age']]),columns=["x"])
```

```python
df3.head(3)
```

```python
df3.describe()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(df3["age"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(df3["age_sc"])
plt.show()
```

## ---- 1.2 Normalization ----

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mc = MinMaxScaler()
mc.fit(df3[["age"]])
```

```python
df3["age_min"] = mc.transform(df3[["age"]])
```

```python
df3.head(3)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(df3["age"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(df3["age_min"])
plt.show()
```
