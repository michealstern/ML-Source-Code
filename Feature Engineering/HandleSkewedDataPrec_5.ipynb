# Log transformations

```python
# Positive skewed ko handle karne ke liye log transformation use hota hai

import numpy as np
import pandas as pd

# Original Data (Sales values)
data = {'Product': ['Product A', 'Product B', 'Product C', 'Product D'],
        'Sales': [1000, 50000, 5000, 200000]}

# Create a DataFrame
df = pd.DataFrame(data)

# Applying Log Transformation (base 10) on Sales column
df['Log_Sales'] = np.log10(df['Sales'])

# Print the transformed DataFrame
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 5))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(df['Sales'])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(df['Log_Sales'])
plt.show()
```

```python
# Sample Data (positively skewed)
data1 = {'Sales': [10, 20, 50, 200, 500, 1000, 5000, 10000]}
df = pd.DataFrame(data1)
df
```

```python
# Apply Log Transformation
df['Log_Sales'] = np.log(df['Sales'])

# Visualize the transformed data
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.title('Before')
sns.histplot(x = 'Sales', data=df)

plt.subplot(1,2,2)
plt.title('After')
sns.histplot(x = 'Log_Sales', data=df)
plt.show()
```

# Square & Cube root transformations

```python
import numpy as np
import pandas as pd

# Sample Data (Sales data with some skewed values)
data = {'Product': ['Product A', 'Product B', 'Product C', 'Product D', 'Product E'],
        'Sales': [100, 1500, 200, 10000, 5000]}

# Create DataFrame
df = pd.DataFrame(data)

df
```

```python
# Apply Square Root Transformation

# Square Root Transformation ka use tab karein jab aapke data mein thoda skewness ho aur aapko data ko compress karna ho.

df['Sqrt_Sales'] = np.sqrt(df['Sales'])

print("Data after Square Root Transformation:")
df
```

```python
# Visualize the transformed data
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.title('Before')
sns.histplot(x = 'Sales', data=df)

plt.subplot(1,2,2)
plt.title('After')
sns.histplot(x = 'Sqrt_Sales', data=df)
plt.show()
```

```python
# Apply Cube Root Transformation

# Cube Root Transformation ka use tab karein jab aapke data mein bohot zyada skewness ho aur aapko extreme values ko zyada compress karna ho.

df['Cbrt_Sales'] = np.cbrt(df['Sales'])

print("Data after Cube Root Transformation:")
df
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.title('Before')
sns.histplot(x = 'Sales', data=df)

plt.subplot(1,2,2)
plt.title('After')
sns.histplot(x = 'Cbrt_Sales', data=df)
plt.show()
```

# polynomial feature transformation

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures

# Sample data (simple linear data)
data = {'X': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)
df
```

```python
# Define Polynomial Features (degree = 2, i.e., X, X)
poly = PolynomialFeatures(degree=2)

# Apply Polynomial Feature Transformation
poly_features = poly.fit_transform(df[['X']])
```

```python
# Convert the result to a DataFrame for better visualization
poly_df = pd.DataFrame(poly_features, columns=['1 (bias)', 'X', 'X^2'])

print("Polynomial Features (degree=2):")
poly_df
```

# Positive and negative Skewed Graph

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate positively skewed data
data_pos = np.random.exponential(scale=2, size=1000)  # Exponential distribution is positively skewed

# Generate negatively skewed data
data_neg = np.random.beta(a=2, b=5, size=1000) * 10  # Beta distribution with a > b is negatively skewed

# Create DataFrames
df_pos = pd.DataFrame(data_pos, columns=['Value'])
df_neg = pd.DataFrame(data_neg, columns=['Value'])

# Plot for Positive Skew
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.distplot(df_pos['Value'], kde=True)
plt.title("Positive Skew (Right Skewed)")

# Plot for Negative Skew
plt.subplot(1, 2, 2)
sns.distplot(df_neg['Value'], kde=True)
plt.title("Negative Skew (Left Skewed)")

plt.show()
```

```python
# Plot for Positive Skew
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df_pos['Value'], kde=True)
plt.title("Positive Skew (Right Skewed)")

# Plot for Negative Skew
plt.subplot(1, 2, 2)
sns.histplot(df_neg['Value'], kde=True)
plt.title("Negative Skew (Left Skewed)")

plt.show()
```

```python
# Calculate skewness
pos_skew = df_pos['Value'].skew()
neg_skew = df_neg['Value'].skew()

print(f"Positive Data Skewness: {pos_skew}")
print(f"Negative Data Skewness: {neg_skew}")
```

```python
# Box plot for Positive Skew
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x=df_pos['Value'])
plt.title("Box Plot of Positive Skewed Data")

# Box plot for Negative Skew
plt.subplot(1, 2, 2)
sns.boxplot(x=df_neg['Value'])
plt.title("Box Plot of Negative Skewed Data")

plt.show()
```

# Differences Between Log, Square Root, and Cube Root Transformations:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a dataset with positive skewness (extreme outliers)
data = [10, 15, 20, 25, 30, 35, 40, 50, 60, 70, 100, 150, 500, 1000, 5000]

# Create a DataFrame
df = pd.DataFrame(data, columns=['Sales'])

# Plotting the original data (before transformation)
plt.figure(figsize=(12, 6))
sns.histplot(df['Sales'], kde=True)
plt.title("Original Sales Data (Positively Skewed)")
plt.show()

# Log Transformation
df['Log_Sales'] = np.log(df['Sales'] + 1)  # Adding 1 to avoid log(0)

# Square Root Transformation
df['Sqrt_Sales'] = np.sqrt(df['Sales'] + 1)  # Adding 1 to avoid sqrt(0)

# Cube Root Transformation
df['CubeRoot_Sales'] = np.cbrt(df['Sales'] + 1)  # Adding 1 to avoid cube root of 0

# Plotting the transformed data

# Log-transformed data
plt.figure(figsize=(12, 6))
sns.histplot(df['Log_Sales'], kde=True)
plt.title("Log-Transformed Sales Data (Strong Compression)")
plt.show()

# Square root transformed data
plt.figure(figsize=(12, 6))
sns.histplot(df['Sqrt_Sales'], kde=True)
plt.title("Square Root-Transformed Sales Data (Moderate Compression)")
plt.show()

# Cube root transformed data
plt.figure(figsize=(12, 6))
sns.histplot(df['CubeRoot_Sales'], kde=True)
plt.title("Cube Root-Transformed Sales Data (Mild Compression)")
plt.show()
```
