```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=1000,noise=0.2)
```

```python
df = {"x1":x[:,0], "x2":x[:,1], "y":y}
```

```python
data = pd.DataFrame(df)
data.head(3)
```

```python
sns.scatterplot(x="x1", y="x2", data=data, hue="y")
plt.show()
```

```python
x_a = data.iloc[:,:-1]
y_a = data["y"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_a,y_a, test_size=0.2, random_state=42)
```

# Bagging

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
sv = SVC()
sv.fit(x_train,y_train)
sv.score(x_test,y_test)*100 , sv.score(x_train,y_train)*100
```

```python
bg = BaggingClassifier(estimator=SVC(),n_estimators=80)
bg.fit(x_train,y_train)
bg.score(x_test,y_test)*100 , bg.score(x_train,y_train)*100
```

# RandomForest

```python
rf = RandomForestClassifier(n_estimators=30)
rf.fit(x_train,y_train)
rf.score(x_test,y_test)*100 , rf.score(x_train,y_train)*100
```
