# Support Vector Machine (SVM)

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Polynomial.csv")
data.head(3)
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Feature1",y="Feature2", data=data, hue="Label")
plt.show
```

```python
x = data.iloc[:,:-1]
y = data["Label"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
{'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'} # 'precomputed' to use must be 0 ,1 columns like One-hot-encodding
```

```python
from sklearn.svm import SVC
sv = SVC(kernel='poly')
sv.fit(x_train,y_train)
sv.score(x_test,y_test)*100 , sv.score(x_train,y_train)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(), clf=sv)
plt.show()
```
