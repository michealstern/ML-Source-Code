```python
import pandas as pd
```

```python
data = pd.read_csv("placement.csv")
data.head(3)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="cgpa", y="resume_score", data=data, hue="placed")
plt.show()
```

```python
# Distribution plot.
plt.figure(figsize=(4,3))
sns.kdeplot(data=data["cgpa"])
plt.show()
```

```python
x = data.iloc[:,:-1]
y = data["placed"]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

# 1. Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gs = GaussianNB()
gs.fit(x_train,y_train)
```

```python
gs.score(x_test,y_test)*100
```

```python
# Decision Regions.
from mlxtend.plotting import plot_decision_regions
```

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(), clf=gs)
plt.show()
```

# 2. Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mn = MultinomialNB()
mn.fit(x_train,y_train)
```

```python
mn.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(), clf=mn)
plt.show()
```

# 3. Bernoulli Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
bn = BernoulliNB()
bn.fit(x_train,y_train)
```

```python
bn.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(5,4))
plot_decision_regions(x.to_numpy(),y.to_numpy(), clf=bn)
plt.show()
```
