# Support Vector Classification

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
data.data # Dataset in dictionary formate.
```

```python
data.feature_names
```

```python
data.target # 0 means yes cancer(malignant tumar), 1 Means No(malignant begin).
```

```python
# Creat DataFrame
df = pd.DataFrame(np.c_[data.data, data.target], columns=[list(data.feature_names)+['target']])
df.head()
```

```python
df.tail()
```

```python
#How many info of patient.(569 patient, 31 Features)
df.shape
```

```python
# Split Data

X = df.iloc[:, 0:-1]  # all rows(:), all columns(0:), leave last columns(-1)
y= df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2020)

print('Shape of X_train = ', X_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_test = ', y_test.shape)
```

# Training Support Vector Classificatin Model

```python
from sklearn.svm import SVC
```

```python
# Train this model.

classification_rbf = SVC(kernel='rbf')
classification_rbf.fit(X_train, y_train)
```

```python
classification_rbf.score(X_test, y_test)
```

```python
# Feature scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_sc = sc.transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
#Feature scale data. check accurecy

classification_rbf_2 = SVC(kernel='rbf')
classification_rbf_2.fit(X_train_sc, y_train)

classification_rbf_2.score(X_test_sc, y_test)
```

# SVC with kernel Polynomial

```python
#Feature scale data in polynomial. check accurecy

classification_poly = SVC(kernel='poly', degree=2)
classification_poly.fit(X_train_sc, y_train)

classification_poly.score(X_test_sc, y_test)
```

# SVC with Kernel Linear

```python
#Feature scale data in Linear. check accurecy

classification_linear = SVC(kernel='linear')
classification_linear.fit(X_train_sc, y_train)

classification_linear.score(X_test_sc, y_test)
```

```python
# Kernel rbf accurecy without feature scaling----> 0.9122807017543859 (Not perfect accurecy)

# Kernel rbf accurecy with feature scaling----> 0.956140350877193   (Not perfect accurecy)

# Kernel Polynomial accurecy with feature scaling----> 0.8333333333333334  (Not perfect accurecy)

# Kernel Linear accurecy with feature scaling----> 0.9649122807017544   (perfect accurecy)
```
