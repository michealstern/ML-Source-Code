```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Social_Network_Ads.csv")
data.head(4)
```

```python
# To check dataset imbalanced or not.
data["Purchased"].value_counts()
```

```python
x = data.iloc[:,:-1]
y = data["Purchased"]
```

```python
# Split Dataset
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
# Model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train ,  y_train)
```

```python
lr.score(x_test , y_test)*100
```

```python
# Predict data..  Actual Value(1), Predicted Value(0)
lr.predict([[45 , 26000]])
```

## ****** Using Random Under Sampling *******

```python
data1 = pd.read_csv("Social_Network_Ads.csv")
data.head(4)
```

```python
# To check dataset imbalanced or not.
data1["Purchased"].value_counts()
```

```python
x1 = data1.iloc[:,:-1]
y1 = data1["Purchased"]
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rs = RandomUnderSampler()
rs_x , rs_y = rs.fit_resample(x1 , y1)
```

```python
rs_y.value_counts()
```

```python
# Split Dataset
from sklearn.model_selection import train_test_split
x_train1 , x_test1 , y_train1 , y_test1 = train_test_split(rs_x,rs_y, test_size=0.2, random_state=42)
```

```python
# Model
from sklearn.linear_model import LogisticRegression
lr1 = LogisticRegression()
lr1.fit(x_train1 ,  y_train1)
```

```python
lr1.score(x_test1 , y_test1)*100
```

```python
# Predict data..  Actual Value(1), Predicted Value(1)
lr1.predict([[48 , 41000]])
```

## ****** Using Random Over Sampling *******

```python
from imblearn.over_sampling import RandomOverSampler
ro = RandomOverSampler()
ro_x , ro_y = ro.fit_resample(x1 , y1)
```

```python
ro_y.value_counts()
```

```python
x_train2 , x_test2 , y_train2 , y_test2 = train_test_split(ro_x,ro_y, test_size=0.2, random_state=42)
```

```python
lr2 = LogisticRegression()
lr2.fit(x_train2 ,  y_train2)
```

```python
lr2.score(x_test2 , y_test2)*100
```

```python
# Predict data..  Actual Value(1), Predicted Value(1)
lr2.predict([[45,26000]])
```
