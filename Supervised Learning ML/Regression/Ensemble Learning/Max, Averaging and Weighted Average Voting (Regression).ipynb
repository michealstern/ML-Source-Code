```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("CGPA-Package.csv")
data.head(3)
```

```python
x = data.iloc[:,:-1]
y = data["package"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)*100 , lr.score(x_train,y_train)*100
```

```python
dr = DecisionTreeRegressor()
dr.fit(x_train,y_train)
dr.score(x_test,y_test)*100 , dr.score(x_train,y_train)*100
```

```python
sv = SVR()
sv.fit(x_train,y_train)
sv.score(x_test,y_test)*100 , sv.score(x_train,y_train)*100
```

# Ensemble Learning

```python
from sklearn.ensemble import VotingRegressor
```

```python
li = [("LinearRegression",LinearRegression(),), ("DecisionTree",DecisionTreeRegressor()), ("SupportVector",SVR())]
```

```python
vr = VotingRegressor(li) # If you want to increase accuracy then do the improve weights=[10,10,10] values.
vr.fit(x_train,y_train)
```

```python
vr.score(x_test,y_test)*100 , vr.score(x_train,y_train)*100
```

# To check base on algorithms answers.

```python
prd = {"LinearRegression":lr.predict(x_test), "DecisionTree":dr.predict(x_test), "SupportVector":sv.predict(x_test),
       "VotingRegressor":vr.predict(x_test)}
```
