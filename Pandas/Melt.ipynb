```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Genre': ['Action','Action','Action','Action','Adventure','Adventure','Comedy','Comedy','Comedy',],
                    'Year': [2008,2009,2008,2010,2009,2010,2010,2008,2009],
                    'Budget(MS)': [28,200,32,20,18,105,20,8,30],
                    'Rating %': [48,63,57,52,84,44,70,81,71]})
df1
```

```python
pd.melt(df1)
```

```python
pd.melt(df1, id_vars = ['Genre'])
```

```python
pd.melt(df1, id_vars = ['Year'])
```

```python
pd.melt(df1, id_vars = ['Year'], value_vars= ['Genre'])
```

```python
pd.melt(df1, id_vars = ['Year'], value_vars= ['Rating %'])
```

```python
pd.melt(df1, id_vars = ['Year'], value_vars= ['Genre'], var_name= 'Category')
```

```python
pd.melt(df1, id_vars = ['Year'], value_vars= ['Genre'], var_name= 'Category', value_name= 'Data')
```
