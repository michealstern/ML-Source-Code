```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'Genre': ['Action','Action','Action','Action','Adventure','Adventure','Comedy','Comedy','Comedy',],
                    'Year': [2008,2009,2008,2010,2009,2010,2010,2008,2009],
                    'Budget(MS)': [28,200,32,20,18,105,20,8,30],
                    'Rating %': [48,63,57,52,84,44,70,81,71]})
df1
```

```python
df1.pivot_table(index = 'Genre')
```

```python
df1.pivot_table(index = 'Genre', columns = 'Year')
```

```python
df1.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'count')
```

```python
df1.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'max')
```

```python
df1.pivot_table(index = 'Genre', columns = 'Year', aggfunc = 'sum')
```

```python
df1.pivot_table(index = 'Genre', columns = 'Year', fill_value = '0')
```

```python
df1.pivot_table(index = 'Genre', columns = 'Year', fill_value = '0', margins= True)
```

```python
df1.pivot_table(index = 'Genre', columns = 'Year', aggfunc='sum', fill_value = '0', margins= True)
```
