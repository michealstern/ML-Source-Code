```python
df1 = pd.DataFrame({'ID': [1,2,3,4], 
              'Name': ['Arman', 'Ruman', 'Noman', 'Rehan'],
              'Class': ['BCA IT', '1st std', '5th std', '7th std']})
df1
```

```python
df2 = pd.DataFrame({'ID': [5,6,7,8], 
              'Name': ['Muskan', 'Noora', 'Kaneez', 'Bushra'],
              'Class': ['Marrid', '10th std', '4th Std', '2nd Std']})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df2, df1])
```

```python
pd.concat([df1, df2], axis = 1)
```

```python
pd.concat([df1, df2], axis = 0, ignore_index = True)
```

```python
df1 = pd.DataFrame({'ID': [1,2,3,4], 
              'Name': ['Arman', 'Ruman', 'Noman', 'Rehan'],
              'Class': ['BCA IT', '1st std', '5th std', '7th std']})
df1
```

```python
df2 = pd.DataFrame({'ID': [5,6], 
              'Name': ['Muskan', 'Noora'],
              'Class': ['Marrid', '10th std']})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 1)
```

```python
pd.concat([df1, df2], axis = 1, join = 'inner')
```

```python
pd.concat([df1, df2], axis = 1, join_axes = [df1.index])
```
